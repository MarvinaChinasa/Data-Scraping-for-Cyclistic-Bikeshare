```python
# Step 4: Define the ResortRoom Class
class ResortRoom:
  # Step 5: The Contructor sets up the room details and its initial empty state
    def __init__(self, room_number, room_type, nightly_rate):
        self.room_number = room_number
        self.room_type = room_type
        self.nightly_rate = nightly_rate
        self.is_booked = False

    # Step 6: Create the Booking Method
    def book_room(self):
        if not self.is_booked:
            self.is_booked = True
            print(f"✨ Success: Room {self.room_number} ({self.room_type}) has been successfully booked!")
        else:
            print(f"❌ Alert: Room {self.room_number} is already occupied. Cannot book.")

    # Step 7: Create the Checkout Method
    def checkout(self):
        if self.is_booked:
            self.is_booked = False
            print(f"🧹 Success: Guest checked out of Room {self.room_number}. Room is now vacant.")
        else:
            print(f"ℹ️ Info: Room {self.room_number} is already empty. No checkout needed.")

    # Step 8: Check Availability Status
    def check_status(self):
        # Translate the True/False boolean into human-readable words
        status_label = "Booked" if self.is_booked else "Available"
        print(f"🏨 Room {self.room_number} ({self.room_type}) — ${self.nightly_rate}/night — Status: {status_label}")
```

```python
# Step 8a. Let's build a specific room instance (Object)
oasis_villa = ResortRoom(101, "Oasis Villa", 850.00)
```

```python
# Step 8b. Check its starting status
print("--- Initial Status Check ---")
oasis_villa.check_status()
```

```python
# Step 8c. Try to book the room for a guest
print("\n--- Processing Booking ---")
oasis_villa.book_room()
```

```python
# Step 8d. Check status again to see if the state changed
print("\n--- Post-Booking Status Check ---")
oasis_villa.check_status()
```

```python
# Step 8e. What happens if we double-book it by accident?
print("\n--- Testing Double-Booking Guard ---")
oasis_villa.book_room()
```

```python
# Step 8f. Guest departs: process checkout
print("\n--- Processing Checkout ---")
oasis_villa.checkout()
```

```python
# Step 8g. Final status verification
print("\n--- Final Status Check ---")
oasis_villa.check_status()
```
